# Order Priority Analysis

```python
df['Order Priority'].value_counts()   #value_count function has returned the count of all unique values in the given index
```

```python
df['Order Priority'].unique()  #unique() method is used to know all type of unique values
```

```python
df['Order Priority']=df['Order Priority'].replace("Critical ","Critical")  #replace() is used to replace a the data from the existing data
```

```python
plt.figure(figsize=(5,4))  #this method is used to change the size of graph
sns.countplot(x="Order Priority",data=df)
plt.title("Count of Order Priority")
plt.savefig("Count of Order Priority.jpg")  #method used to save the given graph
plt.show()
```

# Ship Mode Analysis

```python
df['Ship Mode'].value_counts()
```

```python
x = df['Ship Mode'].value_counts().index
y = df['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(5,4))
plt.pie(y,labels=x,startangle=60,autopct="%0.2f%%")  #autopct is used to tell the percentage in the pie chart
plt.legend()
plt.show
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Ship Mode", data=df,hue="Product Category") #In seaborn, the hue parameter determines which column in the data frame should be used for colour encoding
plt.show
```
